```python
#full code in 1 cell
import pandas as pd
df = pd.read_csv('iris.csv')
setosa_stats = df[df['species'] == 'Iris-setosa'].describe()
versicolor_stats = df[df['species'] == 'Iris-versicolor'].describe()
virginica_stats = df[df['species'] == 'Iris-virginica'].describe()
print("Iris-setosa Stats:\n", setosa_stats)
print("Iris-versicolor Stats:\n", versicolor_stats)
print("Iris-virginica Stats:\n", virginica_stats)
```
